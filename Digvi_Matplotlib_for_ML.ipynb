### Matplotlib

Matplotlib is a low level **graph plotting library in python that serves as a visualization** utility

Matplotlib was created by **John D. Hunter** , it is **open source** and we can **use it freely**.

Matplotlib is **mostly written in pyhton** , a few segments are written in **C , Objective-C and Javascript** for platform compatibility.

```python
# Draw a line from points

import numpy as np
import pandas as pd

xpoints=np.array([0, 6]) # x1=0, x2=6
ypoints=np.array([0, 256]) # y1=0, y2=256
 
plt.plot(xpoints,ypoints) # will show only object
plt.show() # will show graph
```

#### How to use Default X-points

If we do not specify the points in the x-axis, they will get the default values 0, 1, 2, 3 (et. depending on the length of the y-point)

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

If x will not given then plot will take values from default 0 onwards

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

# plt.plot(ypoints, marker= '0')
plt.plot(ypoints, marker= 'd')
plt.show()
```

###### you can also use the shortcut string notation parameter to specify the marker.

#### 1.Marker Size

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

# plt.plot(ypoints, marker = 'o', ms=20)
plt.plot(ypoints, marker='o', ms=25)
plt.show()
```

#### 2. Marker Color

        mec : marker edge color
        mfc : marker face color

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

# plt.plot(ypoints, marker = 'o', ms=20, mec='y')
plt.plot(ypoints, 'o--r', ms=25, mec= 'b')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

# plt.plot(ypoints, marker = 'o', ms=20, mfc='y')
plt.plot(ypoints, 'o--r', ms=25, mfc= 'b')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

# plt.plot(ypoints, marker = 'o', ms=20, mfc='y')
plt.plot(ypoints, 'o--r', ms=25, mfc= 'g')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, 'o--', ms=25, mec= 'b', mfc='y')
plt.show()
```

### Pie Charts

```python
y=np.array([35,25,25,15,55,45])

plt.pie(y)
plt.show()
```

###### labels to pie chart

```python
y=np.array([35,25,25,15])
mylabels=['Apple','Banana','Chery','Mango']
myexplode=[0.1, 0, 0, 0]
plt.pie(y, labels=mylabels, explode=myexplode)
plt.show()
```

###### shadow

```python
y=np.array([35,25,25,15])
mylabels=['Apple','Banana','Chery','Mango']
myexplode=[0.1, 0, 0, 0]
plt.pie(y, labels=mylabels, explode=myexplode, shadow=True)
plt.show()
```

###### colors

```python
y=np.array([35,25,25,15])
mylabels=['Apple','Banana','Chery','Mango']
mycolors=['black','hotpink','b','y']
plt.pie(y, labels=mylabels, colors=mycolors)
plt.show()
```

### Legend

        plt.legend()

```python
y=np.array([35,25,25,15])
mylabels=['Apple','Banana','Chery','Mango']

plt.pie(y, labels=mylabels)
plt.legend(title='Fruits')
plt.show()
```

```python
y=np.array([35,25,25,15])
mylabels=['Apple','Banana','Chery','Mango']

plt.pie(y, labels=mylabels)
plt.legend(title='Fruits',loc=2)
plt.show()
```
